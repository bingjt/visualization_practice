```r
library(reshape2)
library(ggplot2)
library(dplyr)
```

```r
life_expec_pfinal <- life_expec %>%
    group_by(Race, Year) %>%
    filter(Race != "All Races") %>%
    ggplot(aes(x = Year, y = Avg_Life_Expec, group = Race, colour = Race)) + 
    geom_line() + facet_wrap(~Sex, ncol=1)
```
